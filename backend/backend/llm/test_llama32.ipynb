```python
import requests
import json

url = "http://localhost:11434/api/chat"

def llama3(prompt):
    data = {
        "model": "llama3.2",
        "messages": [
            {
                "role": "user",
                "content": prompt

            }
        ],
        "stream": False,
    }

    headers = {
        "Content-Type": "application/json"
    }

    response = requests.post(url, headers=headers, json=data)
    return response.json()
```

```python
response = llama3("who wrote the book godfather")
print(response['message']['content'])
```

```python
# Install the library
# pip install ollama

import ollama

# Basic generation
response = ollama.generate(
    model='llama3.2',
    prompt='What are effective knowledge management techniques?'
)
print(response['response'])

# Chat completion (for conversation)
chat_response = ollama.chat(
    model='llama3.2',
    messages=[
        {
            'role': 'user',
            'content': 'Find connections between cognitive psychology and computer science'
        }
    ]
)
print(chat_response['message']['content'])
```
